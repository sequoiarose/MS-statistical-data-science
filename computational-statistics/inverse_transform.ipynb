# class work

```r
n = 1000
u = runif(n)
results = rep(NA, n)

for (i in 1:n) {
    if (u[i] <= 0.25){
        results[i] = -5
    }
    else if ((0.25 < u[i]) & (u[i]<= 0.75)){
       results[i] = -1
    } 
    else if ((0.75 < u[i] ) & (u[i]<= 0.95)) {
       results[i] = 0
    } 
    else if ((0.95 < u[i]) & (u[i] <= 1)) {
       results[i] = 100
    } 
}
```

```r
barplot((table(results)/n), ylim=c(0,0.6))
```
