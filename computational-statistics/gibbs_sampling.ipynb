# Gibbs Sampling

```r
beta = 2
h = function(m){
    x_max = nrow(m)
    y_max = ncol(m)
    n = 0
    for (i in 1:x_max){
        for (j in 1:y_max){
            last_state = m[i, j]
            if (i<x_max){
                # check down
                n = n +(last_state == m[i + 1, j])
            }
            if (i > 1) {
                # check up
                n = n +(last_state == m[i - 1, j])
            }
            if (j<y_max){
                # check right
                n = n +(last_state == m[i, j+1])
            }
            if (j>1) {
                # check left
                n = n +(last_state == m[i, j-1])
            }
        }
    }
    return(n)
}
```

```r
test = matrix(1, nrow=3, ncol=3)
h(test)
```

```r
gibbs_step = function(current_x){
    for(i in 1:nrow(current_x)){
        for (j in 1:ncol(current_x)){
            current_x[i,j] = 0
            alt_x = current_x
            alt_x[i,j] = 1
            u = runif(1)
            if (u > exp(beta*h(current_x))/(beta*h(current_x)+beta*h(alt_x))){
                current_x = alt_x
            }
        }
    }
    return(current_x)
}
```

```r
# Gibbs sampler

x0 = matrix(rbinom(9, size = 1, p=0.5), nrow=3)

burn_in = 100
runs = 10*burn_in
samples = c()

current_x = x0

for(b in 1:burn_in){
    current_x = gibbs_step(current_x)
}

for(r in 1:runs){
    current_x = gibbs_step(current_x)
    samples[r] = h(current_x)
}
```
