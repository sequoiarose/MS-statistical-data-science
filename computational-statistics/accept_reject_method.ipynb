```r
x = seq(from=0,to=1, 1000)
curve(((3/2)*x^3)+((11/8)*x^2)+((1/6)*x)+1/12, 0, 1)
abline(a=3.125, b=0)
```

```r
# draw from x
x = runif(4500, 0, 1)
#draw u
u = runif(4500, 0, 1)

pi_x <- function(x){
    new_x = ((3/2)*x^3)+((11/8)*x^2)+((1/6)*x)+1/12
    return (new_x)
}

count = 1
accept = c()

while(count <= 4500 & length(accept) < 1000) {
    test_u = u[count]
    test_x = pi_x(x[count])/(3.125*dunif(x[count], 0,1))
    if (test_u <= test_x) {
        accept = rbind(accept, x[count])
        count = count +1
    }
    count = count+1
}

hist(accept, prob=TRUE)
x = seq(from=0,to=1, 1000)
lines(x,(((3/2)*x^3)+((11/8)*x^2)+((1/6)*x)+1/12))
```
