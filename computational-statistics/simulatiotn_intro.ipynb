Steps for a simulation:

1. Plan the set up - e.g., sample size, number of samples
2. Create storage for results - e.g., vector
3. Run the simulation and collect results
4. Look at results - e.g., mean, standard deviation
5. Interpret and analyze the results - e.g., verify it is the expected distribution

## Example

X~Binomial(10,0.2)

$\mu = n*p = 10*0.2=2$

$\sigma^2 = n*p*(1-p) = 10 * 0.2 * 0.8 = 1.6$

$\sigma = 0.4$

Standard error:

$\sigma^2 = \frac{n*p*(1-p)}{100} = \frac{(10 * 0.2 * 0.8)}{100} $

```r
(10*0.2*0.8)/(100)
```

```r
sqrt((10*0.2*0.8)/(100))
```

```r
n = 100
n_samples = 500

vector_of_means = rep(NA, n_samples)

for (i in 1:n_samples){
    obs = rbinom(n, 10, 0.2)
    vector_of_means[i] = mean(obs)
}

print("Mean")

sample_mean = mean(vector_of_means)
print(sample_mean)

print("Standard Deviation")
sample_sd = sd(vector_of_means, na.rm=FALSE)
print(sample_sd)

print("Variance")
print(var(vector_of_means))


hist(vector_of_means, prob=TRUE, ylim=c(0, 3.5))
x = seq( from=1.6, to=2.4, length =100)
lines(x, dnorm(x, mean=sample_mean, sd = sample_sd))
```

## Sampling distribution of median

```r
n = 10
n_samples = 500

vector_of_means = rep(NA, n_samples)

for (i in 1:n_samples){
    obs = rbinom(n, 10, 0.2)
    vector_of_means[i] = median(obs)
}

print("Mean")

sample_mean = mean(vector_of_means)
print(sample_mean)

print("Standard Deviation")
sample_sd = sd(vector_of_means, na.rm=FALSE)
print(sample_sd)

print("Variance")
print(var(vector_of_means))


hist(vector_of_means, prob=TRUE, ylim=c(0, 1))
x = seq( from=0, to=10, length =100)
lines(x, dnorm(x, mean=sample_mean, sd = sample_sd))
```
