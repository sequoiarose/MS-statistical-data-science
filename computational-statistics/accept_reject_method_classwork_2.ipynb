```r
set.seed(1)

n = 5000 # number accept

pi_x <- function(x){
    new_x = dbeta(x, 5, 1)
    return (new_x)
}

count = 1
accept = c()
c = 5.1

while(length(accept) < n) {
    test_u = runif(1, 0, 1) # draw from u
    x = runif(1, 0, 1) # draw x from g(x)
    test_pix_cgx = pi_x(x)/(c*dunif(x, 0,1))
    if (test_u <= test_pix_cgx) {
        accept = rbind(accept, x)
    }
    count = count+1
}

hist(accept, prob=TRUE, breaks=20)
x = seq(from=0, to=1, length.out=1000)
lines(x,  dbeta(x, 5, 1))
```
