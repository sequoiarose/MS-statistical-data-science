```r
set.seed(1)
s = 20000 # number sampled
n = 10000 # number accept
# draw x from g(x)
x = runif(s, 0, 1)
# draw u
u = runif(s, 0, 1)

pi_x <- function(x){
    new_x = dbeta(x, 2, 2)
    return (new_x)
}

count = 1
accept = c()
c = 1.51

while(count <= s & length(accept) < n) {
    test_u = u[count]
    test_pix_cgx = pi_x(x[count])/(c*dunif(x[count], 0,1))
    if (test_u <= test_pix_cgx) {
        accept = rbind(accept, x[count])
    }
    count = count+1
}

hist(accept, prob=TRUE, breaks=20)
x = seq(from=0, to=1, length.out=1000)
lines(x,  dbeta(x, 2, 2))
```
