# Gibbs Classwork

```r
gibbs_step = function(current_x){
    current_x[1] = rnorm(1, current_x[2] / 2, sd = sqrt(3 / 4))
    current_x[2] = rnorm(1, current_x[1] / 2, sd = sqrt(3 / 4))
    return(current_x)
}
```

```r
start =c(1,1)

burn_in = 100
runs = 10*burn_in
samples = matrix(nrow=runs, ncol=2)

current_x = start

for(b in 1:burn_in){
    current_x = gibbs_step(current_x)
}

for(r in 1:runs){
    current_x = gibbs_step(current_x)
    samples[r, 1] = current_x[1]
    samples[r, 2] = current_x[2]
}
```

```r
plot(samples[,1], samples[,2])
```
