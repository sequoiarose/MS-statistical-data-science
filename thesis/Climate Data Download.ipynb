```python
import pandas as pd
import requests 
import os
import datetime
from calendar import monthrange
```

```python
base_url = "https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/county/time-series/CA-"
```

```python
cvars = ["tavg", "tmax", "tmin", "pcp", "cdd", "hdd", "pdsi", "phdi", "pmdi", "zndx"]
headers = [4, 4, 4, 4, 4, 4, 2, 2, 2, 2]
counties = [(2*i)+1 for i in range(0, 58)]
```

```python
sheets = {}
i = 0
for cvar in cvars:
    cvar_data = {}
    # make folder to store downloads 
    os.makedirs(cvar)
    # for each county - create file name, create url, download file
    for county in counties:
        file_name = os.path.join(cvar, str(county)+".csv")
        url = base_url + f"{county:03}" + "/" + cvar + "/all/12/2012-2023.csv?base_prd=true&begbaseyear=1901&endbaseyear=2000"
        print(url)
        r = requests.get(url)  
        with open(file_name, 'wb') as f:
            f.write(r.content)
    # create merged sheet 
    # for file in folder - open file, pull county name, save data cvar_data[county_name] = list of data
    for file in os.listdir(cvar):
        temp_df = pd.read_csv(os.path.join(cvar, file), header=headers[i])
        county = pd.read_csv(os.path.join(cvar, file), nrows=0).columns[0]
        cvar_data[county] = temp_df["Value"].to_list()
    # create pd dataframe from cvar data, save to sheets as sheets[cvar] = df
    ind = [datetime.datetime(int(str(temp_df.at[i,"Date"])[0:4]), int(str(temp_df.at[i,"Date"])[4:]), monthrange(int(str(temp_df.at[i,"Date"])[0:4]), int(str(temp_df.at[i,"Date"])[4:]))[1]) for i in range(len(temp_df))]
    df = pd.DataFrame(cvar_data, index=ind)
    sheets[cvar] = df
    i += 1
# save sheets as one excel file 
with pd.ExcelWriter('climate_data.xlsx') as writer:  
    for sheet in sheets:
        sheets[sheet].to_excel(writer, sheet_name=sheet)
```
